```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import os
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Establishing pathways
executable_path = {'executable_path': ChromeDriverManager().install()}    
browser = Browser('chrome', **executable_path, headless=True)
```

```python
# Defining URL to scrape
news_url = 'https://redplanetscience.com/'
browser.visit(news_url)
```

```python
# Establishing variables for news articles
news_html = browser.html
news_soup = bs(news_html,'lxml')
# print(news_soup)
```

```python
# Extract title and caption text
news_title = news_soup.find("div",class_="content_title").text
news_para = news_soup.find("div", class_="article_teaser_body").text
print(f"Title is: {news_title}")
print(f"Paragraph is: {news_para}")
```

```python
# Mars Hemispheres
mhurl = 'https://www.marshemispheres.com/'
browser.visit(mhurl)
```

```python
# Establishing variables for hemispheres
mhtml = browser.html
print(mhtml)

mh_soup = bs(mhtml,"html.parser")
print(mh_soup)
```

```python
# Selecting the results
results = mh_soup.find_all("div",class_='item')
hemisphere_image_urls = []
for result in results:
        product_dict = {}
        titles = result.find('h3').text
        end_link = result.find("a")["href"]
        image_link = "https://www.marshemispheres.com/" + end_link    
        browser.visit(image_link)
        html = browser.html
        soup= bs(html, "html.parser")
        downloads = soup.find("div", class_="downloads")
        image_url = downloads.find("a")["href"]
        print(titles)
        print(image_url)
        product_dict['title']= titles
        product_dict['image_url']= image_url
        hemisphere_image_urls.append(product_dict)
```
